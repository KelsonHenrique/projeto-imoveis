# 1) Importação e Exploração dos Dados

```python
import pandas as pd
```

Qual o valor médio por tipo de imóvel?

```python
# No Pandas, usamos o método "mean" para calcular a média de um conjunto de dados

dados.Valor.mean()
```

```python
# Saber a média total não é exatamente o que precisamos, pois o que precisamos é dos valores médios por tipo de imóvel.
# Para isso, vamos ter de agrupar os dados para cada tipo de imóvel
# O método "Groupby" é utilizado para essa finalidade
# É interessante observar que o Groupby sempre funciona em conjunto com outro método do Pandas (nesse caso, o "mean")

dados.groupby('Tipo').mean(numeric_only=True) #"numeric_only=True" especifica que esse método válido apenas para colunas numéricas
```

```python
dados.groupby('Tipo').mean()
```

```python
pd.__version__ # Saber a versão do Pandas 
```

```python
dados.groupby('Tipo').Valor.mean()  

# Colocando a variável "Valor" antes do "mean", eu estou estabelecendo que a média seja calculada apenas para o "Valor"
```

```python
# Para tornar a apresentação desses dados mais atraente, vamos transformá-los num DataFrame

dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor') 
```

```python
# Outra forma de transformar seria...
# Obs: Nessa forma, não preciso colocar o argumento na função "sort_values", pois ele vale apenas para Series

pd.DataFrame(dados.groupby('Tipo').Valor.mean().sort_values())
```

```python
# Vamos salvar esse DataFrame em uma variável

df_preço_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor') 
```

# Demanda do time de Dev

O time de desenvolvedores do site da empresa solicitou a criação de duas novas colunas numéricas na nossa base de dados. No site dos imóveis, eles gostariam de apresentar informações mais sumarizadas em relação a valores. Para isso, essas colunas serão importantes:

valor_por_mes: essa coluna deve conter os gastos mensais de cada imóvel, incluindo aluguel e condomínio;

valor_por_ano: essa coluna deve conter os gastos anuais por imóvel, ou seja, IPTU mais 12 meses de aluguel e condomínio.

```python
url2 = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
```

```python
data = pd.read_csv(url,sep=';') # Para o site da imobiliária, as informações de todos os imóveis são importante
```

```python
data.head()
```

```python
# Vamos criar a coluna "Valor por mês"

data['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
```

```python
data.head()
```

```python
# Criando a coluna 'Valor por ano'

data['Valor_por_ano'] = data['Valor_por_mes'] * 12 + data['IPTU']
```

```python
data.head()
```

O time de desenvolvimento solicitou a criação de mais duas colunas. No entanto, dessa vez elas são categóricas:

Descrição: essa coluna deve possuir uma sumarização das principais informações dos imóveis que serão apresentadas no site: tipo de imóvel, bairro, quantidade de quartos e vagas de garagem;

Possui_suite: essa deve ser uma coluna que informe apenas se o imóvel possui ou não suítes, sem se importar com a quantidade.

```python
data['Descriçao'] = data['Tipo'] + ' em ' + data['Bairro']
```

```python
data.head()
```

```python
data['Descriçao'] = data['Tipo'] + ' em ' + data['Bairro'] + ' com ' + data['Quartos'] + 'quarto(s)' + \
' e ' + data['Vagas'] + 'vaga(s) de garagem.'
```

```python
# Esse erro aconteceu, pois só consigo concatenar string com string. E eu estava concatenando string com inteiro
```

```python
data['Descriçao'] = data['Tipo'] + ' em ' + data['Bairro'] + ' com ' + data['Quartos'].astype(str) + 'quarto(s)' + \
' e ' + data['Vagas'].astype(str) + 'vaga(s) de garagem.'
```

```python
data.head()
```

```python
data['Possui_suite'] = data['Suites'].apply(lambda x: 'Sim' if x > 0 else "Não")

# O "apply" é um método utilizado para aplicar uma determinada função num conjunto específico de dados
```

```python
data.head()
```
